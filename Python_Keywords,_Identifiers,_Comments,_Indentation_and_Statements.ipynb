```python
# 2) Describe the rules for defining identifiers in Python and provide an example.'''

'''
Identifiers are user-defined names given to variables, functions, classes, modules, and other objects.
They play a crucial role in making code more readable and understandable.
The rules for defining identifiers are follows:

1) Reserve Keywords cannot be used as identifiers.
2) Identifiers can contain Letters (lowercase and uppercase), Digits, Underscore.
3) They cannot start with a digit.
4) Identifiers can start with an underscore.
5) There is no strict limit on identifier length.
6) Identifiers are case-sensitive.

Example:-

# Valid Identifiers:
   o  ab10c (Contains only letters and numbers)
   o  abc_DE (valid characters)
   o  _abc (Identifiers can start with an underscore)

#Invalid Identifiers:
   o  99 (Cannot be only digits)
   o  9abc (Cannot start with a number)
   o  x+y (Only an underscore is allowed as a special character)
   0 for (Reserved keyword is not allowed)

'''
```

```python
# 4)  Why is proper indentation important in Python?

'''
Proper indentation is crucial in Python because it significantly impacts code
readability, structure, and functionality. Let’s delve into why it matters:

1) Code Readability:
   o  Indentation breaks up long blocks of code, making it easier to visually identify logical sections.
   o  Well-indented code is more readable and maintainable.
   o  It helps developers quickly understand the program’s flow.

2) Block Identification:
   o  Indentation defines blocks (such as loops, conditionals, functions, etc.).
   o  Each indented block represents a specific purpose or scope.
   o  Proper indentation ensures that statements within the same block are grouped together.

3) Syntax and Errors:
   o  Incorrect indentation leads to IndentationErrors during code execution.
   o  Python relies on consistent indentation to determine block boundaries.
   o  Mismatched indentation can cause unexpected behavior or syntax issues.

4) Visual Hierarchy:
   o  Indentation visually organizes code hierarchically.
   o  Nested blocks are indented further to the right.
   o  It helps in understand the nesting level of loops, conditionals, and function definitions.

   '''
```

```python
# 6) Differentiate between expression and statement in Python with examples.

'''
The differentiate between expressions and statements in Python:

Expressions:

1) An expression is any combination of symbols, words, or values that evaluates to a single value.
2) Expressions can be assigned to variables or used as operands in operations.
3) They are unique in meaning and always result in a value.

Examples of expressions:
- 2 + 3 (evaluates to 5)
- len("hello") (evaluates to the length of the string)
- x * y (where x and y are variables)


Statements:

1) A statement is a complete line of code that performs an action or task.
2) Statements can be composed of one or more expressions.
3) They can have side effects (e.g., modifying variables, printing output).

Examples of statements:
- Assignment statement: x = 10
- Conditional statement (if-else): if condition: ... else: ...
- Loop statement (for, while): for item in iterable: ...
- Function definition: def my_function(): ...


There are few Key Differences:-
   o  Expressions can be assigned or used as operands, while statements can only be declared.
   o  Statements create side effects to be useful, while expressions are values or execute to values.
   o  Expressions are unique in meaning, while statements are two-sided (they either do tasks or don’t).

Here are some examples:
Expression:
x = 2 + 3  # The expression 2 + 3 evaluates to 5 and is assigned to x

Statement:
if temperature > 30:
    print("It's a hot day!")  # The whole line is a statement

'''
```
